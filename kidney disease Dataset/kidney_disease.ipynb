***type conversion***

into a float

```python
df['packed_cell_volume'].isna().sum()  ## 70 null values ----------------------
```

```python
df['packed_cell_volume'] = pd.to_numeric(df['packed_cell_volume'] , errors='coerce')
```

```python
df.info()
```

```python
df['white_blood_cell_count'] = pd.to_numeric(df['white_blood_cell_count'] , errors='coerce')
```

```python
df['red_blood_cell_count'] = pd.to_numeric(df['red_blood_cell_count'] , errors='coerce')
```

```python
df.info()
```

```python
cat_cols = [col for col in df.columns if df[col].dtype == "object"]
cat_cols
```

```python
num_cols = [col for col in df.columns if df[col].dtype != "object"]
num_cols
```

```python
for col in cat_cols:
    print(f'{col} has {df[col].unique()} values \n')
```

```python
#replace unwanted values

df['diabetes_mellitus'].replace(to_replace={' yes' : 'yes' , '\tyes' : 'yes' , '\tno' : 'no'}, inplace=True)
```

```python
df['diabetes_mellitus'].unique()
```

```python
df['coronary_artery_disease'].replace(to_replace={'\tno' : 'no'}, inplace=True)
```

```python
df['coronary_artery_disease'].unique()
```

```python
df['class'].replace(to_replace={ 'ckd\t': 'ckd', 'notckd\t': 'notckd'}, inplace=True)
```

```python
df['class'].unique()
```

```python
df['class'] = df['class'].map({'ckd': 0, 'notckd': 1})
```

```python
df['class'].unique()
```

```python
#Univariate

plt.figure(figsize = (10, 6))
sns.histplot(df['age'].dropna(), kde = True, bins = 20, color = 'skyblue')
plt.title('Dist of age')
plt.xlabel('Age')
plt.ylabel('frequency')
plt.show()
```
